# **1.A**

```
import numpy as np

def yex(t):
  return 0.2*t*np.exp(3*t) - 0.04*np.exp(3*t) +0.04*np.exp(-2*t)  #initial
def f(t,y):
  return t*np.exp(3*t) - 2*y  #euler

a = 0
b = 1
N = 10
alpha = 0.5

#step1
h = (b-a)/N
t = a
w = alpha
wplot =np.array([alpha])
yt = yex(t)
yplot = np.array([yt])
print("t  w yex |w-yex|")
print("{} {} {} {}".format(t,w ,yt, np.abs(w-yt)))

#step 2
for i in range(1,N +1):
  w = w + h*f(t,w) #step3
  wplot = np.append(wplot, w)
  t = a + i*h #step4
  yt = yex(t)
  yplot = np.append(yplot,yt)
  print("{} {} {} {}".format(t,w ,yt, np.abs(w-yt)))

import matplotlib.pyplot as plt

tex = np.linspace(0,2,11)
plt.grid()
plt.plot(tex, wplot)
plt.plot(tex, yplot)
plt.legend(['w-Euler','y-exact'], fontsize =15)
plt.show()
```
